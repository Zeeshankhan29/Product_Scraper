```python
from flask import Flask, render_template, request,jsonify
from flask_cors import CORS,cross_origin
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
```

```python
search=input('Enter the input to search',)
No_of_pages=int(input('Enter the no of pages to extract the information'))


product_name=[]
product_description=[]
product_price=[]
product_ratings=[]
product_image=[] 

for i in range(1,No_of_pages+1,1):
    url='https://www.flipkart.com/search?q='
    flipkart_page=requests.get(url+search+"&page"+str(i))
    flipkart_html=bs(flipkart_page.text,'html.parser')
    products=flipkart_html.findAll('div',{"class":"_4rR01T"})
    Product_description=flipkart_html.findAll('ul',{"class":"_1xgFaf"})
    Product_price=flipkart_html.findAll('div',{"class":"_30jeq3 _1_WHN1"})
    Ratings=flipkart_html.findAll('div',{"class":"_3LWZlK"})[0:25]
    image_url=flipkart_html.findAll('div',{"class":"_2QcLo-"})
  

    
    for i in range(1,24):
        try:
            products[i].text==[]
            print(products[i].text)
            product_name.append(products[i].text)

        except:
            print('There is no such html tags')
        try:
            Product_description[i].text==[]
            print(Product_description[i].text)
            product_description.append(Product_description[i].text)

        except:
            print('There is no such html tags')
        try:
            Product_price[i].text==[]
            print(Product_price[i].text)
            product_price.append(Product_price[i].text)

        except:
            print('There is no such html tags')
        try:
            Ratings[0:25][i].text==[]
            print(Ratings[0:25][i].text)
            product_ratings.append(Ratings[0:25][i].text)

        except:
            print('There is no such html tags')
        try:
            image_url[i].div.div.img['src']==[]
            print(image_url[i].div.div.img['src'])
            product_image.append(image_url[i].div.div.img['src'])

        except:
            print('There is no such html tags')
        print('\n')

```

```python
import pandas as pd
```

```python
df=pd.DataFrame({"Product_name": product_name, "Product_Description": product_description,
                     'Product_price': product_price, 'Product_ratings': product_ratings, "Image": product_image})
```
